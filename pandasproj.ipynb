```python
#replace missing bill amount with median bill amount of respective department
df['bill_amount']=df.groupby('department')['bill_amount'].transform(lambda x: x.fillna(x.median()))
```

```python
print(df['bill_amount'].isnull().sum())
```

```python
#handiling missing doctor_name with mode
df['doctor_assigned']=df['doctor_assigned'].fillna('Unknown')
```

```python
#fix datatypes
df['admission_date']=pd.to_datetime(df['admission_date'])
```

```python
#consistant spacing
cat_cols=['gender','department','outcome','treatment_type','insurance_status']
for col in cat_cols:
    df[col]=df[col].str.strip().str.title()
```

```python
pd.to_datetime(df['admission_date'])
```

```python
df["admission_date"] = pd.to_datetime(df["admission_date"])
df["discharge_date"] = pd.to_datetime(df["discharge_date"])

df["stay_days"] = (df["discharge_date"] - df["admission_date"]).dt.days
```

```python
print(df[['admission_date','discharge_date','stay_days']].head())
```

```python
df['age_group']=pd.cut(df['age'],bins=(0,15,30,45,60),labels=['Child','Youth','Adult','Senior'])
print(df[['age','age_group']].head())
```

```python
df["cost_category"] = pd.cut(
    df["bill_amount"],
    bins=[0,30000,70000,120000,300000],
    labels=["Low","Medium","High","Very High"]
)
print(df[['bill_amount','cost_category']].head())   
```

```python
df['readmission_risk']=df['readmission_flag'].map({0:'No',1:'yes'})
print(df[['readmission_flag','readmission_risk']].head())
# Export cleaned data
```

```python
cleaned_data=pd.read_csv(r"c:\Users\Hp\Downloads\hospital_data_cleaned.csv")
```

```python
#Exploratory Data Analysis (EDA) & KPIs
#overall dataset health
cleaned_data.info()
cleaned_data.describe()
```

```python
#length of stay
cleaned_data["discharge_date"] = pd.to_datetime(cleaned_data["discharge_date"])
cleaned_data["admission_date"] = pd.to_datetime(cleaned_data["admission_date"])
cleaned_data['length_of_stay'] = (cleaned_data["discharge_date"] - cleaned_data["admission_date"]).dt.days
print(cleaned_data["length_of_stay"])
```

```python
cleaned_data['length_of_stay']=pd.to_numeric(cleaned_data['length_of_stay'],errors='coerce')
cleaned_data['bill_amount']=pd.to_numeric(cleaned_data['bill_amount'],errors='coerce')
cleaned_data['readmission_risk']=cleaned_data['readmission_risk'].astype('category')
```

```python
dept_kpi=cleaned_data.groupby('department').agg(patients=('patient_id','count'),
                                               avg_bill=('bill_amount','mean'),
                                               avg_length_of_stay=('length_of_stay','mean')).sort_values('patients',ascending=False)
```

```python
print(dept_kpi)
```

```python
cost_counts = cleaned_data["cost_category"].value_counts()
cost_percent = cleaned_data["cost_category"].value_counts(normalize=True) * 100

print("Counts:\n", cost_counts)
print("\nPercentages:\n", cost_percent)
```
